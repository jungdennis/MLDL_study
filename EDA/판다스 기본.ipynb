# 5. Concat

```python
# 위아래로 concat
df_1 = np.array([[1, 3, 5], [2, 4, 6]])
df_1 = pd.DataFrame(df_1, columns = ["A", "B", "C"])
df_2 = -df_1

df_3 = pd.concat([df_1, df_2])
df_3
```

```python
# 양 옆으로 concat
df_3 = pd.concat([df_1, df_2], axis = 1)
df_3
```

# 7. Missing Values

```python
df_C = pd.read_excel("toy.xlsx")
```

```python
# 결측치가 하나라도 있는 행 모두 제거하기
df_C.dropna()
```

```python
# 결측치 채우기 - 같은 열 앞 행의 값
df_C.fillna(method = "ffill")
```

```python
# 결측치 채우기 - 해당 열의 평균값
df_C.fillna(df_C.mean())
```

```python
# 결측치 채우기 - 전체 평균값
df_C.fillna(df_C.mean().mean())
```

# 8. Aggregating

```python
# 새로운 열 추가하기
df = pd.DataFrame([[1, -1, 2], [2, -2, 3], [3, -3, 4], [4, -4, 7], [5, -5, 8]], columns = ["A", "B", "C"])
df["평균"] = df.mean(axis = 1)
df
```

```python
# 새로운 행 추가하기
df.loc[5] = df.sum()
df
```

```python
# 특정 column을 기준으로 묶어서 합계 구하기
df_2 = pd.DataFrame({"val1" : [1, 2, 3, 4, 5], "val2" : [-1, -2, -3, -4, -5],
                     "val3" : [2, 3, 4, 7, 8], "label" : ["A", "A", "B", "B", "B"]})
df_2.groupby("label").sum()
```

# 9. Pivot Table

```python
# pivot table 만들기
df_2.pivot_table(index = "label", aggfunc = "mean")
```
